```python
#Analysis on home ownership
home_analysis = loan_data.pivot_table(index="home_ownership",columns="loan_status",aggfunc="count",fill_value=0,values="id")
home_analysis.reset_index(inplace = True)
home_analysis
```

```python
home_analysis.plot(x="home_ownership",kind="bar")
```

```python
def calculator(df):
    total = df["Charged Off"]+df["Current"]+df["Fully Paid"]
    df["Charged Off"]= (df["Charged Off"]/total)*100
    df["Current"]= (df["Current"]/total)*100
    df["Fully Paid"]= (df["Fully Paid"]/total)*100
    return df
    
```

```python
calculator(home_analysis)
home_analysis
```

```python
home_analysis.plot(x="home_ownership",kind="bar")
```
